````python
### Task 2.2: Slicing operations (8 points)
import numpy as np

# -----------------------------
# 1D Array Slicing
# -----------------------------

# Create array from 0 to 19
arr = np.arange(0, 20)

# 1. Extract elements from index 5 to 15 (inclusive)
# Upper bound is exclusive, so we use 16
slice_5_15 = arr[5:16]

# 2. Extract every 3rd element from the array
every_third = arr[::3]

# 3. Reverse the entire array using slicing
reversed_arr = arr[::-1]

print("Original array:", arr)
print("Elements from index 5 to 15:", slice_5_15)
print("Every 3rd element:", every_third)
print("Reversed array:", reversed_arr)


# -----------------------------
# 2D Array Slicing
# -----------------------------

# Create a 4x4 2D array
arr_2d = np.arange(1, 17).reshape(4, 4)

print("\n2D Array:\n", arr_2d)

# Extract the first 2 rows and first 3 columns
#===============I use AI for how to extract 2 rows and 3 columns===========
first_rows_cols = arr_2d[:2, :3]

# Extract the last 2 rows and last 2 columns
last_rows_cols = arr_2d[-2:, -2:]

# Extract every other row with all columns
alternate_rows = arr_2d[::2, :]

print("\nFirst 2 rows and first 3 columns:\n", first_rows_cols)
print("\nLast 2 rows and last 2 columns:\n", last_rows_cols)
print("\nEvery other row (all columns):\n", alternate_rows)
````

````python
### Task 2.3: Boolean and fancy indexing (7 points)
import numpy as np

# Given array
arr = np.array([15, 23, 8, 42, 11, 37, 19, 5, 33, 28])

print("Original array:", arr)

# 1. Extract all values greater than 20 using boolean indexing
greater_than_20 = arr[arr > 20]

# 2. Extract all even numbers
# Modulo (%) operator checks divisibility by 2
even_numbers = arr[arr % 2 == 0]

# 3. Extract values between 10 and 30 (inclusive)
# Combine multiple conditions using logical AND (&)
between_10_30 = arr[(arr >= 10) & (arr <= 30)]

# 4. Fancy indexing: extract elements at indices 0, 3, 5, and 8
indices = [0, 3, 5, 8]
fancy_elements = arr[indices]

# Print results
print("Values greater than 20:", greater_than_20)
print("Even numbers:", even_numbers)
print("Values between 10 and 30:", between_10_30)
print("Fancy indexing result:", fancy_elements)
````

````python
## Exercise 4: Array Manipulation (20 points)

### Task 4.1: Reshaping and transposing (10 points)

#Create an array with values from 1 to 12, then:
#1. Reshape it to 3x4
#2. Reshape it to 2x6
#3. Reshape it to 2x2x3 (3D array)
#4. Transpose the 3x4 array
#5. Flatten the 3x4 array back to 1D
import numpy as np

# Create a 1D array with values from 1 to 12
arr = np.arange(1, 13)

print("Original 1D array:", arr)

# 1. Reshape the array to 3x4
arr_3x4 = arr.reshape(3, 4)
print("\nReshaped to 3x4:\n", arr_3x4)

# 2. Reshape the array to 2x6
arr_2x6 = arr.reshape(2, 6)
print("\nReshaped to 2x6:\n", arr_2x6)

# 3. Reshape the array to 2x2x3 (3D array)
arr_3d = arr.reshape(2, 2, 3)
print("\nReshaped to 2x2x3 (3D array):\n", arr_3d)

# 4. Transpose the 3x4 array (swap rows and columns)
#======================I use AI for tronspose function================
transposed = arr_3x4.T
print("\nTransposed 3x4 array:\n", transposed)

# 5. Flatten the 3x4 array back to 1D
flattened = arr_3x4.flatten()
print("\nFlattened array:", flattened)
````

````python
### Task 4.2: Stacking and splitting (10 points)

#Create three 1D arrays: `[1, 2, 3]`, `[4, 5, 6]`, `[7, 8, 9]`

#1. Stack them vertically to create a 3x3 array
#2. Stack them horizontally to create a 1D array

#Create an array from 1 to 12 and:
#- Split it into 4 equal parts

#Create a 4x4 array with values 1-16 and:
#- Split it horizontally into 2 parts
import numpy as np

# -----------------------------
# Stacking 1D arrays
# -----------------------------

# Create three 1D arrays
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.array([7, 8, 9])
#============Use AI for stack operations=======
# 1. Stack vertically to create a 3x3 array
vertical_stack = np.vstack((a, b, c))

# 2. Stack horizontally to create a 1D array
horizontal_stack = np.hstack((a, b, c))

print("Vertical stack (3x3):\n", vertical_stack)
print("\nHorizontal stack (1D):", horizontal_stack)


# -----------------------------
# Splitting 1D array
# -----------------------------

# Create an array from 1 to 12
arr = np.arange(1, 13)

# Split into 4 equal parts
split_1d = np.split(arr, 4)

print("\nSplit 1D array into 4 equal parts:")
for part in split_1d:
    print(part)


# -----------------------------
# Splitting 2D array
# -----------------------------

# Create a 4x4 array with values from 1 to 16
arr_2d = np.arange(1, 17).reshape(4, 4)

# Split horizontally into 2 parts (along columns)
horizontal_split = np.hsplit(arr_2d, 2)

print("\nOriginal 4x4 array:\n", arr_2d)
print("\nHorizontal split into 2 parts:")
for part in horizontal_split:
    print(part)
````

````python
## Exercise 5: Statistical Operations (15 points)

### Task 5.1: Basic statistics (7 points)

#Given `arr = np.array([12, 45, 23, 67, 34, 89, 56, 78, 90, 31])`:

#Calculate and print:
#1. Sum of all elements
#2. Mean (average)
#3. Median
#4. Standard deviation
#5. Variance
#6. Minimum value
#7. Maximum value
#8. Index of minimum value
#9. Index of maximum value

#solution
import numpy as np

# Given array
arr = np.array([12, 45, 23, 67, 34, 89, 56, 78, 90, 31])

print("Array:", arr)

# Sum of all elements
total_sum = np.sum(arr)

#  Mean (average)
mean_value = np.mean(arr)

#  Median
median_value = np.median(arr)

# Standard deviation
std_deviation = np.std(arr)

#  Variance
variance_value = np.var(arr)

#  Minimum value
min_value = np.min(arr)

#  Maximum value
max_value = np.max(arr)

#  Index of minimum value
min_index = np.argmin(arr)

#  Index of maximum value
max_index = np.argmax(arr)

# Print results
print("Sum:", total_sum)
print("Mean:", mean_value)
print("Median:", median_value)
print("Standard Deviation:", std_deviation)
print("Variance:", variance_value)
print("Minimum value:", min_value)
print("Maximum value:", max_value)
print("Index of minimum value:", min_index)
print("Index of maximum value:", max_index)

````

````python
# ## Exercise 6: Real-World Application - Student Grade Analysis (25 points)

# You are given grade data for 5 students across 4 subjects (Math, Science, English, History):

# ```python
# grades = np.array([
#     [85, 90, 78, 88],
#     [92, 88, 95, 90],
#     [76, 82, 80, 75],
#     [88, 85, 90, 92],
#     [95, 92, 88, 94]
# ])
# ```

# **Tasks:**

# 1. **Calculate the average grade for each student** (5 points)
#    - Output should show Student 1: X.XX, Student 2: Y.YY, etc.

# 2. **Calculate the average grade for each subject** (5 points)
#    - Output: Math: X.XX, Science: Y.YY, etc.

# 3. **Find the highest and lowest grades in each subject** (5 points)
#    - Output for each subject: Highest = X, Lowest = Y

# 4. **Identify students with an average grade above 85** (3 points)
#    - Print student number and their average

# 5. **Calculate the overall class average** (2 points)
#    - Average of all grades across all students

# 6. **Normalize all grades to a scale of 0-10** (5 points)
#    - Use min-max normalization formula: (value - min) / (max - min) * 10
#    - Print the normalized grades array

# **Bonus:** Identify the top-performing student (+2 points)

# **Expected Output Format:**
# ```
# --- Task 1: Student Averages ---
# Student 1: XX.XX
# ...
#
# --- Task 2: Subject Averages ---
# Math: XX.XX
# ...
#
# (Continue for all tasks)
# ```

# Solution
import numpy as np

# Given grades array
grades = np.array([
    [85, 90, 78, 88],
    [92, 88, 95, 90],
    [76, 82, 80, 75],
    [88, 85, 90, 92],
    [95, 92, 88, 94]
])

subjects = ["Math", "Science", "English", "History"]

# --------------------------------------------------
# Task 1: Average grade for each student
# --------------------------------------------------
student_averages = np.mean(grades, axis=1)

print("--- Task 1: Student Averages ---")
for i, avg in enumerate(student_averages, start=1):
    print(f"Student {i}: {avg:.2f}")


# --------------------------------------------------
# Task 2: Average grade for each subject
# --------------------------------------------------
subject_averages = np.mean(grades, axis=0)

print("\n--- Task 2: Subject Averages ---")
for subject, avg in zip(subjects, subject_averages):
    print(f"{subject}: {avg:.2f}")


# --------------------------------------------------
# Task 3: Highest and lowest grades in each subject
# --------------------------------------------------
subject_max = np.max(grades, axis=0)
subject_min = np.min(grades, axis=0)

print("\n--- Task 3: Highest and Lowest Grades per Subject ---")
for i, subject in enumerate(subjects):
    print(f"{subject}: Highest = {subject_max[i]}, Lowest = {subject_min[i]}")


# --------------------------------------------------
# Task 4: Students with average above 85
# --------------------------------------------------
print("\n--- Task 4: Students with Average Above 85 ---")
for i, avg in enumerate(student_averages, start=1):
    if avg > 85:
        print(f"Student {i}: {avg:.2f}")


# --------------------------------------------------
# Task 5: Overall class average
# --------------------------------------------------
overall_average = np.mean(grades)

print("\n--- Task 5: Overall Class Average ---")
print(f"Overall Average: {overall_average:.2f}")


# --------------------------------------------------
# Task 6: Normalize grades to scale 0–10
# --------------------------------------------------
min_grade = np.min(grades)
max_grade = np.max(grades)

normalized_grades = (grades - min_grade) / (max_grade - min_grade) * 10

print("\n--- Task 6: Normalized Grades (0–10 scale) ---")
print(np.round(normalized_grades, 2))


# --------------------------------------------------
# Bonus: Top-performing student
# --------------------------------------------------
top_student_index = np.argmax(student_averages)

print("\n--- Bonus: Top-Performing Student ---")
print(f"Student {top_student_index + 1} with average {student_averages[top_student_index]:.2f}")
````

````python
# ## Exercise 7: Challenge - Sales Data Analysis (30 points)

# You are analyzing sales data for a company with 4 products over 6 months:

# ```python
# sales_data = np.array([
#     [120, 135, 150, 145, 160, 175],  # Product A
#     [90, 95, 100, 110, 115, 120],    # Product B
#     [150, 145, 155, 160, 170, 180],  # Product C
#     [80, 85, 82, 88, 92, 95]         # Product D
# ])
# ```

# **Tasks:**

# 1. **Calculate total sales for each product** (5 points)
#    - Sum across all months for each product

# 2. **Calculate total sales for each month** (5 points)
#    - Sum across all products for each month (Jan to Jun)

# 3. **Identify the month with the highest total sales** (4 points)
#    - Print the month name and total sales

# 4. **Calculate the growth rate for each product** (6 points)
#    - Growth rate = ((Last month - First month) / First month) × 100
#    - Print as percentage for each product

# 5. **Find products with average monthly sales greater than 120k** (4 points)
#    - Calculate average for each product
#    - Print products meeting the criteria

# 6. **Calculate quarter-wise sales** (6 points)
#    - Q1: January-March
#    - Q2: April-June
#    - Print sales for each product in each quarter

# **Bonus:** Predict next month (July) sales for each product (+5 points)
# - Use a simple method: average of last 2 months + average monthly growth

# **Your Code:**
# ```python
# import numpy as np
#
# sales_data = np.array([
#     [120, 135, 150, 145, 160, 175],
#     [90, 95, 100, 110, 115, 120],
#     [150, 145, 155, 160, 170, 180],
#     [80, 85, 82, 88, 92, 95]
# ])
# ```
#Solution

# In this code i used AI for
import numpy as np

# Sales data for 4 products over 6 months (Jan to Jun)
# Rows -> Products (A, B, C, D)
# Columns -> Months (Jan, Feb, Mar, Apr, May, Jun)
sales_data = np.array([
    [120, 135, 150, 145, 160, 175],  # Product A
    [90, 95, 100, 110, 115, 120],    # Product B
    [150, 145, 155, 160, 170, 180],  # Product C
    [80, 85, 82, 88, 92, 95]         # Product D
])

products = ["Product A", "Product B", "Product C", "Product D"]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]

# --------------------------------------------------
# Task 1: Total sales for each product
# --------------------------------------------------
product_total_sales = np.sum(sales_data, axis=1)

print("--- Task 1: Total Sales per Product ---")
for product, total in zip(products, product_total_sales):
    print(f"{product}: {total}k")


# --------------------------------------------------
# Task 2: Total sales for each month
# --------------------------------------------------
monthly_total_sales = np.sum(sales_data, axis=0)

print("\n--- Task 2: Total Sales per Month ---")
for month, total in zip(months, monthly_total_sales):
    print(f"{month}: {total}k")


# --------------------------------------------------
# Task 3: Month with highest total sales
# --------------------------------------------------
max_month_index = np.argmax(monthly_total_sales)

print("\n--- Task 3: Highest Sales Month ---")
print(f"{months[max_month_index]}: {monthly_total_sales[max_month_index]}k")


# --------------------------------------------------
# Task 4: Growth rate for each product
# Growth = ((Last - First) / First) * 100
# --------------------------------------------------
growth_rate = ((sales_data[:, -1] - sales_data[:, 0]) / sales_data[:, 0]) * 100

print("\n--- Task 4: Growth Rate per Product ---")
for product, rate in zip(products, growth_rate):   #======I used AI for function=======
    print(f"{product}: {rate:.2f}%")


# --------------------------------------------------
# Task 5: Products with average monthly sales > 120k
# --------------------------------------------------
average_sales = np.mean(sales_data, axis=1)

print("\n--- Task 5: Products with Average Monthly Sales > 120k ---")
for product, avg in zip(products, average_sales):
    if avg > 120:
        print(f"{product}: {avg:.2f}k")


# --------------------------------------------------
# Task 6: Quarter-wise sales
# Q1 -> Jan, Feb, Mar
# Q2 -> Apr, May, Jun
# --------------------------------------------------
q1_sales = np.sum(sales_data[:, :3], axis=1)
q2_sales = np.sum(sales_data[:, 3:], axis=1)

print("\n--- Task 6: Quarter-wise Sales ---")
for i, product in enumerate(products):
    print(f"{product} - Q1: {q1_sales[i]}k, Q2: {q2_sales[i]}k")


# --------------------------------------------------
# Bonus: Predict July sales
# Method:
# July = average of last 2 months + average monthly growth
# --------------------------------------------------
last_two_month_avg = np.mean(sales_data[:, -2:], axis=1)
monthly_growth_avg = (sales_data[:, -1] - sales_data[:, 0]) / 5

july_prediction = last_two_month_avg + monthly_growth_avg

print("\n--- Bonus: Predicted July Sales ---")
for product, prediction in zip(products, july_prediction):
    print(f"{product}: {prediction:.2f}k")
````
