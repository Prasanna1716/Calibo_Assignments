Task 3.2: Implement Missing Value Handling
Create a copy of the original DataFrame called df_clean and implement your strategy:

```python
import pandas as pd
import numpy as np

# Create a copy of the original DataFrame
df_clean = df.copy()

# 1. Drop rows with missing customer_id
df_clean = df_clean.dropna(subset=['customer_id'])

# 2. Drop rows with missing customer_name
df_clean = df_clean.dropna(subset=['customer_name'])

# 3. Drop rows with missing email
df_clean = df_clean.dropna(subset=['email'])

# 4. Fill missing phone numbers with a constant value
df_clean['phone'] = df_clean['phone'].fillna('Not Available')

# 5. Handle registration_date using forward fill then backward fill
df_clean['registration_date'] = df_clean['registration_date'].ffill().bfill()

# 6. Fill missing age values with median
df_clean['age'] = df_clean['age'].fillna(df_clean['age'].median())

# 7. Fill missing purchase_amount with median
df_clean['purchase_amount'] = df_clean['purchase_amount'].fillna(
    df_clean['purchase_amount'].median()
)

# 8. Fill missing product_category with mode
df_clean['product_category'] = df_clean['product_category'].fillna(
    df_clean['product_category'].mode()[0]
)

# 9. Fill missing payment_method with mode
df_clean['payment_method'] = df_clean['payment_method'].fillna(
    df_clean['payment_method'].mode()[0]
)

# 10. (Optional numeric interpolation if required by question)
df_clean[['age', 'purchase_amount']] = df_clean[['age', 'purchase_amount']].interpolate()

# Verify missing values after cleaning
print("Missing values after handling:")
print(df_clean.isnull().sum())

# Display final shape
print("\nFinal shape of df_clean:", df_clean.shape)
```

Task 7.2: Handle Purchase Amount Outliers

Use the IQR (Interquartile Range) method to:

Calculate Q1, Q3, and IQR

Determine lower and upper bounds

Identify outliers

Handle outliers by capping them at the bounds

```python
import numpy as np

#  Calculate Q1 and Q3
Q1 = df_clean['purchase_amount'].quantile(0.25)
Q3 = df_clean['purchase_amount'].quantile(0.75)

#  Calculate IQR
IQR = Q3 - Q1

#  Determine lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Q1 (25th percentile):", Q1)
print("Q3 (75th percentile):", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

# Identify outliers
outliers = df_clean[
    (df_clean['purchase_amount'] < lower_bound) |
    (df_clean['purchase_amount'] > upper_bound)
]

print("\nOutlier values before handling:")
display(outliers[['purchase_amount']])

#  Handle outliers by capping at the bounds
df_clean['purchase_amount'] = np.where(
    df_clean['purchase_amount'] < lower_bound,
    lower_bound,
    np.where(
        df_clean['purchase_amount'] > upper_bound,
        upper_bound,
        df_clean['purchase_amount']
    )
)

# Verify outliers after capping
outliers_after = df_clean[
    (df_clean['purchase_amount'] < lower_bound) |
    (df_clean['purchase_amount'] > upper_bound)
]

print("\nOutliers after handling:")
display(outliers_after[['purchase_amount']])

# Show new range of purchase_amount
print("\nNew purchase_amount range after capping:")
print("Minimum:", df_clean['purchase_amount'].min())
print("Maximum:", df_clean['purchase_amount'].max())
#1.How many outliers were identified?

#The number of outliers identified is equal to the number of records displayed in the outliers table before handling. These values lie outside the lower and upper bounds calculated using the IQR method.

#2. What were the outlier values?

#The outlier values are the purchase_amount values shown in the “Outlier values before handling” output. These values were either below the lower bound or above the upper bound, making them extreme compared to the rest of the data.

#3. What is the new range of purchase amounts?

#After handling outliers by capping:

#The minimum purchase amount is equal to the lower bound

#The maximum purchase amount is equal to the upper bound
```

Challenge 3: Advanced Outlier Detection

Implement Z-score method for outlier detection and compare results with IQR method.


Formula: Z-score = (value - mean) / standard_deviation

Threshold: |Z-score| > 3

```python
import pandas as pd
import numpy as np

# Load cleaned dataset
df_clean = pd.read_csv("/content/cleaned_customer_data.csv")

# -------------------------------
# Z-SCORE METHOD
# -------------------------------

# Calculate mean and standard deviation
mean_pa = df_clean['purchase_amount'].mean()
std_pa = df_clean['purchase_amount'].std()

# Calculate Z-scores
df_clean['z_score'] = (df_clean['purchase_amount'] - mean_pa) / std_pa

# Identify Z-score outliers (|Z| > 3)
zscore_outliers = df_clean[abs(df_clean['z_score']) > 3]

print("🔹 Z-score Outlier Detection")
print("Mean:", mean_pa)
print("Standard Deviation:", std_pa)
print("Number of Z-score outliers:", zscore_outliers.shape[0])

print("\nZ-score outlier values:")
display(zscore_outliers[['purchase_amount', 'z_score']])

# -------------------------------
# IQR METHOD (for comparison)
# -------------------------------

Q1 = df_clean['purchase_amount'].quantile(0.25)
Q3 = df_clean['purchase_amount'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

iqr_outliers = df_clean[
    (df_clean['purchase_amount'] < lower_bound) |
    (df_clean['purchase_amount'] > upper_bound)
]

print("\n🔹 IQR Outlier Detection")# prints IQR Outlier Detection
print("Lower Bound:", lower_bound)#prints Lower Bound
print("Upper Bound:", upper_bound)#prints Upper Bound
print("Number of IQR outliers:", iqr_outliers.shape[0])

print("\nIQR outlier values:")
display(iqr_outliers[['purchase_amount']])
```
